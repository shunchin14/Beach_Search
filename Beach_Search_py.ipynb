```python
# スクレイピングしたcsvのReviewを読み込み
import pandas as pd
data = pd.read_csv('/content/Scraping - Beach_n_Review.csv')
review_list = data['Review'].values.tolist()
print(review_list)
print(type(review_list))
```

```python
# Mecabのインストール、インポート
!pip install mecab-python3 unidic-lite
import MeCab

# 分かち書き
wakati = MeCab.Tagger("-Owakati")
wakati.parse('') # バグ回避
```

```python
# 取り出したい品詞
select_conditions = ['形容詞', '名詞']
terms_list = []

def wakati_text(review) :
  # 分かち書きしてノードごとにする
  node = wakati.parseToNode(review)
  terms = []

  while node :

    # 単語
    term = node.surface
    # 品詞
    pos = node.feature.split(',')[0]

    # もし品詞が条件と一致してたら
    if pos in select_conditions :
      terms.append(term)
    
    # 次のノードへ
    node = node.next

  return terms

wakati_result = [wakati_text(review) for review in review_list]

print(wakati_result)
print(type(wakati_result))
print(len(wakati_result))
```

```python
# categoryのcsvをlistで読み込む
import csv

def read_csv() :
  csv_file = open('/content/category_list.csv')
  csv_reader = csv.reader(csv_file)
    
  category_list = []

  for row in csv_reader :
    if csv_reader.line_num == 1 :
      continue
    category_list.append(row[0])

  csv_file.close()
  return category_list

print(read_csv())
print(type(read_csv()))
print(len(read_csv()))
```

```python
# TF-IDFの定義
from math import log

N = len(wakati_result)
 
def tf(t, d) :
  return d.count(t)/len(d)
 
def idf(t) :
  df = 0
  for doc in wakati_result :
    df += t in doc
    try :
      return log(N/df)+1
    except ZeroDivisionError :
      return 1
 
def tfidf(t, d) :
  return tf(t,d)* idf(t)
```

```python
# TF
result_tf = []
for i in range(N) :
  result_tf.append([])
  d = wakati_result[i]
  for j in range(len(read_csv())):
    t = read_csv()[j]
        
    result_tf[-1].append(tf(t,d))
        
tf_ = pd.DataFrame(result_tf, columns = read_csv())
```

```python
# IDF
result_idf = []
for j in range(len(read_csv())) :
  t = read_csv()[j]
  result_idf.append(idf(t))

idf_ = pd.DataFrame(result_idf, index = read_csv(), columns = ["IDF"])
```

```python
# TF-IDF
result = []
for i in range(N):
    result.append([])
    d = wakati_result[i]
    for j in range(len(read_csv())):
        t = read_csv()[j]
        
        result[-1].append(tfidf(t,d))

tfidf_ = pd.DataFrame(result, columns=read_csv())
```

```python
# Beachの名前とTF-IDFをひとつのDataFrameに
beach_tfidf = pd.concat([data['Beach'], tfidf_], axis = 1)
beach_tfidf
```

## **TF-IDF + Word2Vec**

```python
# BeachごとにTF-IDFの値が高い単語を出力
charas = []
for i in range(len(review_list)) :
  tfidf_dict = tfidf_.iloc[i].to_dict()
  sort_tfidf_dict = sorted(tfidf_dict.items(), key = lambda x:x[1], reverse = True)[:20]
  chara = []
  for j in range(len(sort_tfidf_dict)) :
    chara.append(sort_tfidf_dict[j][0])
  charas.append(chara)

beach_chara = pd.DataFrame(data['Beach'])
beach_chara['character'] = charas
beach_chara
```

```python
# Word2Vecをインストール、インポート
!pip install gensim
from gensim.models import Word2Vec
```

```python
# モデル構築
model = Word2Vec(wakati_result, sg = 1, size = 100, window = 5, min_count = 3, iter = 100, workers = 2)
print(model)
```

```python
import numpy as np

# コサイン類似度を算出する関数を定義
def cos_sim(v1, v2) :
    return np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))
```

```python
# 夕日
avg_scores_sunset = []
for i in range(len(beach_chara['character'])) :
  scores_sunset = []
  for j in range(len(beach_chara['character'][i])) :
    score_sunset = cos_sim(model.wv['夕日'], model.wv[beach_chara['character'][i][j]]) # '夕日'と各単語間のコサイン類似度
    scores_sunset.append(score_sunset)
  avg_scores_sunset.append(np.mean(scores_sunset)) # 20個の単語間スコアの平均値

# beach_charaに列を追加
beach_chara['avg_score_sunset'] = avg_scores_sunset

# 平均値に合わせて並べ替えて上位5つのビーチを表示
vec_sunset = beach_chara.sort_values(by = 'avg_score_sunset', ascending = False)[:5]
vec_sunset['Beach']
```
